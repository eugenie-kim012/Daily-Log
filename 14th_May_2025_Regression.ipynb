## 라이브러리 불러오기

대표적인 선형 회귀 값 (310 Pg)
* 일반 선형 회귀; 예측값과 실제값의 RSS(Residual Sum of Squares)를 최소화 할 수 있도록 회귀 계수를 최적화하며, Regulations을 적용 하지 않는 모델
* 릿지 (Ridge): 선형 회귀에 L2 규제를 추가한 회귀 모델. L2 규재는 상대적으로 큰 회귀 계수 값의 예측 영향도를 감소시키기 위해서 회귀 계수값을 더 작게 만드는 규제 모델
* 라쏘 (Lasso): 예측 영향력이 작은 피쳐의 회귀 계수를 0으로 만들어 회귀 예측시 피처가 선택되지 않게 하는 것 - L1 규제는 피처 선택 기능이라고도 불림
* 엘라스틱넷 (ElasticNet):L2규제와 L1 규제를 결합한 모델.
* 로지스틱 회귀 (Logistic Regression) - 분류에 사용되는 회귀모델. 일반적으로 이진 분류 뿐만 아니라 희소 영역의 분류 (텍스트 분류)에 쓰임

최적의 회귀 모델 - 전체 데이터의 잔차 (오류값) 합이 최소가 되는 모델을 만든다 (일반적으로 오류의 값을 정확히 구하고 미분도 편하게 하기 위해 RSS의 방식을 취함)

# 즉, RSS을 최소로 하는 회귀 계수를 학습을 통해 찾는 것이 머신러닝 기반 회귀의 핵심 사항. 이 떄, 회귀 계수 W이 RSS의 중심 변수임을 이해하기

* RSS - 비용 (Cost),
* RSS: 비용함수
* 머신러닝 회귀 알고리즘 - 데이터를 계속 학습하면서 이 비용 함수가 반환하는 값, 즉 오류값을 지속해서 감소시키고 최종적으로는 더 이상 감소하지 않는 최소의 오류값을 구하는 것
* 비용 함수 (Cost Function) - 손실 함수 (Loss Function)


경사 하강법 (Gradient Descent) - 점진적으로 반복적인 계산을 통해 W 파라미터 값을 업데이터 하면서 오류 값이 최소가 되도록 한다.


```python
# 라이브러리 불러오기
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error #회계 모델을 평가하기 위하여 (실제값과 오차값 차이를 계산하기 위함)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

np.random.seed(0)
# y = 4X + 6 식을 근사(w1=4, w0=6). random 값은 Noise를 위해 만듬
X = 2 * np.random.rand(100,1) # np.random.rand(100,1)은 0과 1 사이의 균등분포 난수 100개를 1열짜리 벡터로 생성
y = 6 +4 * X+ np.random.randn(100,1)

# X, y 데이터 셋 scatter plot으로 시각화
plt.scatter(X, y)
```

## 데이터 불러오기

```python
# 데이터 생성
from sklearn.datasets import load_diabetes # 당뇨병 환자 데이터

def make_dataset():
    dataset = load_diabetes()
    df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
    df['target'] = dataset.target
    X_train, X_test, y_train, y_test = train_test_split(
        df.drop('target', axis=1), df['target'], test_size=0.2, random_state=1004)
    return X_train, X_test, y_train, y_test

X_train, X_test, y_train, y_test = make_dataset()
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# train 샘플
X_train.head()
```

```python
#x타겟 확인 - 연속된 숫자로 구성되어 있는 것을 확인
y_train.head()
```

```python
y_train.hist()
```

```python
# 선형 회귀
from sklearn.linear_model import LinearRegression
model = LinearRegression() #변수 안에 ㅠ넣기
model.fit(X_train, y_train)
pred = model.predict(X_test)
mean_squared_error(y_test, pred)
```

## 릿지 회귀

릿지 회귀
이번 시간에는 릿지 회귀(Ridge regression)에 대해 설명해 드리도록 하겠습니다.

이번 시간 정리
1. 릿지 회귀

오버피팅 문제로 규제(regularization)를 적용한 모델
규제 방식: L2 규제
파라미터(alpha) 값을 활용해 조절
alpha 값이 커질수록 회귀 계수 값을 작게 만듦
*규제를 통해서 모델의 성능을 올릴 수 있습니다.

image.png

```python
# 릿지 회귀 - 선형 회귀 모델모다 오차 (MEAN ERROR)가 커짐, 이 경우 좋지 않은 모델이라는 의미임
from sklearn.linear_model import Ridge
model = Ridge(alpha=1)
model.fit(X_train, y_train)
pred = model.predict(X_test)
mean_squared_error(y_test, pred)
```

```python
# 회귀 계수
coef = pd.DataFrame(data=model.coef_, index=X_train.columns, columns=['alpha1'])
# 컬럼 이름 적는 것, 컬럼의 기본 알파 값은 1임 (규제를 조절하는 alpha 값 기억하기)
coef #변수에 분석의 값 저장하기
```

```python
# 릿지 회귀 (alpha=10) - alpha 값을 10으로 조정하는 것
model = Ridge(alpha=10)
model.fit(X_train, y_train)
pred = model.predict(X_test)
mean_squared_error(y_test, pred)
```

```python
# 회귀 계수 (alpha 10)
coef['alpha10'] = model.coef_
coef
#규정을 통해서 회계 계수를 조정하기
```

```python
# 릿지 회귀 (alpha=0.05)
model = Ridge(alpha=0.05)
model.fit(X_train, y_train)
pred = model.predict(X_test)
mean_squared_error(y_test, pred)
```

```python
# 회귀 계수 (alpha 0.05) - 저장 하고 비교하기 - 규제에 따라 숫자가 달라지는 것 비교하기
coef['alpha0.05'] = model.coef_
coef
```

라쏘 회귀

라쏘 회귀
이번 시간에는 라쏘 회귀(Lasso regression) 모델에 대해 살펴보도록 하겠습니다.

이번 시간 정리
1. 라쏘 회귀

오버피팅 문제로 규제(regularization)를 적용한 모델
규제 방식: L1 규제 (중요한 피처만 선택하는 특성)
L2에 비해 회귀 계수를 급격히 감소시켜 중요하다고 생각하는 피처만 선택하고 나머지는 0으로 만듦
[리마인드] 규제 방식 비교

릿지 회귀 : L2 규제 방식
라쏘 회귀 : L1 규제 방식 (중요한 피처만 선택하는 특성)

image.png

```python
# 라쏘 회귀
from sklearn.linear_model import Lasso
model = Lasso(alpha=1) #기본 알파값이 1
model.fit(X_train, y_train)
pred = model.predict(X_test)
mean_squared_error(y_test, pred)
```

```python
# 회귀 계수, 비교를 진행하기 위해 - 리지 회귀와 비교하면서 분석하기 (중요하지 않다고 생각하는 건 0으로 만들어 버리는 성질이 있음)
coef = pd.DataFrame(data=model.coef_, index=X_train.columns, columns=['alpha1'])
coef
```

```python
# 라쏘 회귀 (alpha=2)
model = Lasso(alpha=2)
model.fit(X_train, y_train)
pred = model.predict(X_test)
mean_squared_error(y_test, pred)
```

```python
# 회귀 계수 (alpha=2)
coef['alpha2'] = model.coef_
coef
```

```python
# 라쏘 회귀 (alpha=0.05)
model = Lasso(alpha=0.05)
model.fit(X_train, y_train)
pred = model.predict(X_test)
mean_squared_error(y_test, pred)
```

```python
# 회귀 계수 (alpha=0.05)
coef['alpha0.05'] = model.coef_
coef
```

## 엘라스틱넷 회귀

엘라스틱넷 회귀
이번 시간에는 엘라스틱넷(ElasticNet) 회귀 모델에 대해 살펴보도록 하겠습니다.

이번 시간 정리
1. 엘라스틱넷 회귀

오버피팅 문제로 규제(regularization)를 적용한 모델
규제 방식: L2 규제와 L1 규제 (중요한 피처만 선택하는 특성) 결합

릿지 회귀 : L2 규제 방식
라쏘 회귀 : L1 규제 방식 (중요한 피처만 선택하는 특성)
엘라스틱넷 회귀 : L2+L1 규제 결합 방식

image.png

```python
# 엘라스틱넷 회귀
from sklearn.linear_model import ElasticNet
model = ElasticNet(alpha=1) # 역시 기본값은 알파 1
model.fit(X_train, y_train)
pred = model.predict(X_test)
mean_squared_error(y_test, pred)
```

```python
# 엘라스틱넷 회귀
from sklearn.linear_model import ElasticNet
model = ElasticNet(alpha=0.0001, l1_ratio=0.6) #알파 값을 조정하였음
model.fit(X_train, y_train)
pred = model.predict(X_test)
mean_squared_error(y_test, pred)
```

## 하이퍼파라미터 튜닝

이번 시간 정리
모델의 성능 향상을 위해 하이퍼파라미터 튜닝을 진행하며 사이킷런의 model_selection에서 제공하는 GridSearchCV와 RandomizedSearchCV를 이용하여 최적화할 수 있습니다.

1. GridSearchCV

grid search를 통해 최적의 하이퍼파라미터를 찾음 (시간 오래 걸림)
2. RandomizedSearchCV

RandomizedSearchCV 는 GridSearchCV와 달리 랜덤으로 N개의 조합만 탐색함
(시간 내 최적의 하이퍼파라미터 탐색)


```python
# 라이브러리 불러오기 (관련된 라이브러리 확인)
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
# 하이퍼파라미터
# 시간이 오래 걸려 learning_rate와 subsample는 주석 처리함
params={# 'learning_rate':[0.07, 0.05],
        'max_depth':[3, 5, 7],
        'n_estimators':[100, 200],
        # 'subsample':[0.9, 0.8, 0.7]
        }
```

```python
# 데이터셋 로드
def make_dataset2():
    dataset = load_diabetes()
    df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
    df['target'] = dataset.target
    return df.drop('target', axis=1), df['target']
X, y = make_dataset2()
```

```python
# GridSearchCV
#  'max_depth':[3, 5, 7],  'n_estimators':[100, 200]만 적용 시 3분 내외
xgb = XGBRegressor()
grid = GridSearchCV(xgb, params, cv=3, n_jobs=-1)
grid.fit(X, y)
```

```python
# 최적의 하이퍼파라미터를 찾음
grid.best_params_
```

```python
# 하이퍼파라미터 튜닝
xgb = XGBRegressor(
    learning_rate = 0.05,
    max_depth = 3,
    n_estimators = 100,
    subsample =0.7
)
xgb.fit(X_train, y_train)
pred = xgb.predict(X_test)
mean_squared_error(y_test, pred)
```

```python
#  Randomized Search
xgb = XGBRegressor()
grid = RandomizedSearchCV(xgb, params, cv=3, n_iter=10, n_jobs=-1)
grid.fit(X, y)
```

```python
# 최적의 하이퍼파라미터를 찾음
grid.best_params_
```
